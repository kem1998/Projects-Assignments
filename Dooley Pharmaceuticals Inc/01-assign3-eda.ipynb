# 8.0 Deviding monthly income into quartiles and inserting the new column

```python
# Deviding the monthly income column into quartile ranges
# less than 4747.5 = Q1, between 4747.5-9495 = Q1-Q2, between 9495-14242.5 = Q2-Q3 and between 14242.5-19999 = Q3-Q4
# Define a function mip_category, where mipr = monthly income quratile range
# Pass the whole dataframe as parameter
# And return a category
def mipr_category(df): 
    if df["monthly_income"] <= 4747.5:
        return "Q1"
    elif (df["monthly_income"] > 4747.5) & (df["monthly_income"] <= 9495):
        return "Q1-Q2"
    elif (df["monthly_income"] > 9495) & (df["monthly_income"] <= 14242.5):
        return "Q2-Q3"
    elif (df["monthly_income"] > 14242.5) & (df["monthly_income"] <= 19999):
        return "Q3-Q4"

# Use apply() to apply a lambda function for all the rows of the dataframe
# The lambda function calls the previously defined function mip_category()
employeeattrition_df["monthly_income_quartile_category"] = \
    employeeattrition_df.apply(lambda employeeattrition_df:mipr_category(employeeattrition_df), axis = 1) # axis=1 = row-wise
# Check
employeeattrition_df.head()
```

```python
# Separate the data in the attrition column to employee retention and attrition
empunsatisfaction_df = employeeattrition_df[employeeattrition_df["job_satisfaction"]<=2]
empsatisfaction_df = employeeattrition_df[employeeattrition_df["job_satisfaction"]>=3]
```

```python
# Ploting the charts using Seaborn's countplot()

# AS 2 separate plots side by side
fig, ax = plt.subplots(1,2)
ax1 = sns.countplot(x="monthly_income_quartile_category", data=empunsatisfaction_df, ax=ax[1]) # employee attrition dataset
ax2 = sns.countplot(x="monthly_income_quartile_category", data=empsatisfaction_df, ax=ax[0]) # employee retention dataset


ax1.set_title("employee Unsatisfaction", fontsize=25)
ax1.set_xlabel("Monthly Income Quartile Range", fontsize=25)
ax2.set_title("employee Satisfaction", fontsize=25)
ax2.set_xlabel("Monthly Income Quartle Range", fontsize=25)
ax1.tick_params(axis='both', which='both', labelsize=30) # x/y axes, minor/major ticks for ax1
ax2.tick_params(axis='both', which='both', labelsize=30) # x/y axes, minor/major ticks for ax2


# If you want to save the graph to disk
plt.tight_layout() # automatically adjusts subplot to fit in the figure area
plt.savefig('monthly_income_quartile_range_category-side-by-side.png', dpi=300) # dpi=300 is print quality
```
